```python
var='atemp2m'
#coords 
#point 1: lon -3.5, lat 40.5
min_lon = -4
min_lat = 40
max_lon = -3
max_lat = 41
#point 2: lon 14.5, lat 57.5
#min_lon = 14
#min_lat = 57
#max_lon = 15
#max_lat = 58

obs_array=np.zeros([24,4,2,24]) # nyears, nmonths, ndays (only two for the example),hours
exp_array=np.zeros([24,4,2,4,25]) # nyears, nmonths, ndays (only two for the example),hours,nmembers
for imonth,month in enumerate(range(5,9)):
    for iyear,year in enumerate(range(1993,2017)):
        for iday,day in enumerate(range(14,16)):
            if var=='atemp2m':
                obs=xr.open_dataset('/work/csp/vt17420/ERA5/hourly/atemp2m/atemp2m_r360x180/'+var+'_'+str(year)+str(month).zfill(2)+'.nc')
            else:
                obs=xr.open_dataset('/data/csp/vt17420/CLINT_proj/ERA5/hourly/'+var+'_r360x180/'+var+'_'+str(year)+str(month).zfill(2)+'.nc')
            obs=obs.sel(time=obs.time[(obs.time.time.dt.month==month)&(obs.time.time.dt.day==day)])
            obs1 = obs.sel(lat=slice(min_lat,max_lat),lon=slice(min_lon,max_lon))
            obs_array[iyear,imonth,iday,:]=obs1.to_array()[0,:,0,0]
            exp=xr.open_dataset('/data/csp/vt17420/CLINT_proj/C3S/meteo_france-7/6hourly/'+var+'_r360x180/Europe/'+var+'_'+str(year)+'0501.nc')
            exp1=exp.sel(time=exp.time[(exp.time.time.dt.month==month)&(exp.time.time.dt.day==day)])
            if var=='atemp2m':
                exp1=exp1.sel(latitude=slice(min_lat,max_lat),longitude=slice(min_lon,max_lon))
            else:
                exp1=exp1.sel(lat=slice(min_lat,max_lat),lon=slice(min_lon,max_lon))
            exp_array[iyear,imonth,iday,:,:]=exp1.to_array()[0,:,:,0,0]
```

```python
# we add the time step of the next day in the same dimension to have the 6-hourly values at 0,6,12,18,24 UTC
exp_array_fake=np.zeros([24,4,1,5,25])
for imonth,month in enumerate(range(5,9)):
    for iyear,year in enumerate(range(1993,2017)):
        for iday,day in enumerate(range(1,2)):
            exp_array_fake[iyear,imonth,iday,0:4,:]= exp_array[iyear,imonth,iday,:,:]
            exp_array_fake[iyear,imonth,iday,4,:]= exp_array[iyear,imonth,iday+1,0,:]  
```

```python
# climatology
obs_mean=np.mean(obs_array,axis=0) 
exp_mean=np.mean(exp_array_fake,axis=0)
exp_em_mean=np.mean(exp_mean,axis=3)
```

```python
obstime= obs_mean[0,0,:].ravel()
y=obstime[0:24]
x=np.arange(0,24 ,1)
z=np.polyfit(x, y,4)
p = np.poly1d(z)
```

```python
mymodel=np.poly1d(np.polyfit([0,6,12,18,24], exp_em_mean[0,0,:]-273.15,4))
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 6))
plt.plot(y-273.15,color='blue',label="ERA5 t2m")
plt.plot(y-273.15,'ro',color='blue')
plt.plot(p(x)-273.15,'b--',label="ERA5 t2m fitted")
plt.plot([0,6,12,18],exp_em_mean[0,0,0:4]-273.15,'ro',color='black',label='C3S t2m')
plt.xlim([-0.5, 23.5])
ax.axvspan(21, 24, alpha=0.2 ,facecolor='0.5',label='(23-06) local time')
ax.axvspan(-0.5,4, alpha=0.2, facecolor='0.5')
plt.plot(mymodel(x),'r--',color='black',label='C3S t2m fitted')
plt.xlabel("timestep (UTC)",fontsize=18)
plt.ylabel("Apparent Temperature (°C)",fontsize=18)
ax.grid(linestyle='--')
ax.tick_params(labelsize=18)
plt.legend(bbox_to_anchor=(1, 0.5),fontsize=14)
ax.set_title('Apparent Temperature at 2m ERA5 climatological daily cycle (1993-2016) \n for the 15th of May in 40.5° N and -3.5° E',fontsize=18)
plt.savefig('/work/csp/vt17420/daily_cycle_app_Spain.png',bbox_inches="tight", dpi=300)
```
